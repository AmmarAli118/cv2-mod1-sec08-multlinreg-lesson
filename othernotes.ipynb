```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

from sklearn.datasets import make_regression

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_set = make_regression(n_samples=1000, n_features=1, n_informative=1, noise=12)
x = data_set[0]
y = data_set[1]
plt.scatter(x, y)
```

```python
linreg = sm.OLS(y, x).fit()
linreg.summary()
```

```python
y_hat = linreg.params[0]*x
y_hat = y_hat.flatten()
```

```python
plt.scatter(x, y)
plt.plot(x, y_hat, c='r', linestyle='--', linewidth=3)
plt.show()
```

```python
residuals = y_hat - y
```

```python
plt.hist(residuals)
```

```python
plt.scatter(np.linspace(0,1,len(residuals)), residuals)
plt.ylim(-40,40)
plt.hlines(y=0,xmin=0, xmax=1, colors='r')
plt.show()
```
